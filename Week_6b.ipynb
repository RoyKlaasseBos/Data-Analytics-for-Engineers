### 1. Determine the movie titles with the biggest loss (worldwide gross - production budget) by genre of which the movie language was not French, German or Spanish and where the loss was at least 10 million dollar.

```python
### STEP BY STEP SOLUTION
```

Determine the movie titles with the **biggest loss (worldwide gross - production budget)** by genre of which the movie language was not French, German or Spanish and where the loss was at least 10 million dollar.

```python
# calcualte profit / loss per movie
query1 = '''SELECT movie_title, \"Worldwide Gross\" - \"Production Budget\"
            FROM movies;    
'''

pd.read_sql(query1, conn).head(10) # only show first 10 records
```

Determine the movie titles with the biggest loss (worldwide gross - production budget) **by genre of which the movie language was not French, German or Spanish** and where the loss was at least 10 million dollar.

```python
# exclude French, German, Spanish movies 
query1 = '''SELECT movie_title, \"Worldwide Gross\" - \"Production Budget\", language
            FROM movies
            WHERE (language != "French") AND (language != "German") AND (language != "Spanish");
'''

pd.read_sql(query1, conn).head(10) #Star Wars Ep. VII: The Force Awakens is now excluded (why?)
```

```python
# exclude French, German, Spanish movies using NOT IN (preferred way - why?)
query1 = '''SELECT movie_title, \"Worldwide Gross\" - \"Production Budget\"
            FROM movies
            WHERE language NOT IN ("French", "German", "Spanish");           
'''

pd.read_sql(query1, conn).head(10)
```

Determine the movie titles with the biggest loss (worldwide gross - production budget) **by genre** of which the movie language was not French, German or Spanish and where the loss was at least 10 million dollar.

```python
# select movie with biggest losss per genre
query1 = '''SELECT movie_title, genre, MIN(\"Worldwide Gross\" - \"Production Budget\") 
            FROM movies
            WHERE language NOT IN ("French", "German", "Spanish")
            GROUP BY genre;
'''
pd.read_sql(query1, conn)
```

Determine the movie titles with the biggest loss (worldwide gross - production budget) by genre of which the movie language was not French, German or Spanish and where **the loss was at least 10 million dollar.**

```python
# only look at genres of which the loss is greater than 10M
query1 = '''SELECT movie_title, genre, MIN(\"Worldwide Gross\" - \"Production Budget\")
            FROM movies
            WHERE language NOT IN ("French", "German", "Spanish")
            GROUP BY genre
            HAVING MIN(\"Worldwide Gross\" - \"Production Budget\") < -10000000;
''' 

pd.read_sql(query1, conn) # why is there a minus symbol in front of 10M? which genre did we exclude? 
```

```python
# change column name to profit
query1 = '''SELECT movie_title, genre, MIN(\"Worldwide Gross\" - \"Production Budget\") AS profit
            FROM movies
            WHERE language NOT IN ("French", "German", "Spanish")
            GROUP BY genre
            HAVING MIN(\"Worldwide Gross\" - \"Production Budget\") < -10000000;
'''

pd.read_sql(query1, conn)
```

### 2. What are the top 3 movies released in 2016 which were not profitable but got the highest IMDB-score?

```python
### STEP BY STEP SOLUTION
```

What are the top 3 **movies released in 2016** which were not profitable but got the highest IMDB-score?

```python
query2 = '''SELECT *
            FROM movies
            WHERE \"Release Date\" >= '2016-01-01' AND \"Release Date\" < '2017-01-01';
'''
pd.read_sql(query2, conn).head() # how can we easily check whether we did this correctly?
```

```python
# find latest / most recent dates
query2 = '''SELECT *
            FROM movies
            WHERE \"Release Date\" >= '2016-01-01' AND \"Release Date\" < '2017-01-01'
            ORDER BY \"Release Date\" DESC;
'''
pd.read_sql(query2, conn).head() # order ascending/descending
```

What are the top 3 movies released in 2016 which were **not profitable** but got the highest IMDB-score?

```python
query2 = '''SELECT movie_title, \"Release Date\", \"Worldwide Gross\" - \"Production Budget\"
           FROM movies
           WHERE \"Release Date\" >= '2016-01-01' AND \"Release Date\" < '2017-01-01' AND \"Worldwide Gross\" - \"Production Budget\" < 0;

'''
pd.read_sql(query2, conn).head()
```

```python
# we don't need to see the profit (i.e. loss), neither the release date
query2 = '''SELECT movie_title
           FROM movies
           WHERE \"Release Date\" >= '2016-01-01' AND \"Release Date\" < '2017-01-01' AND \"Worldwide Gross\" - \"Production Budget\" < 0; 
'''
pd.read_sql(query2, conn).head()
```

```python
# we do need the imdb_score column though
query2 = '''SELECT movie_title, imdb_score
           FROM movies
           WHERE \"Release Date\" >= '2016-01-01' AND \"Release Date\" < '2017-01-01' AND \"Worldwide Gross\" - \"Production Budget\" < 0;
'''
pd.read_sql(query2, conn).head()
```

What are the top 3 movies released in 2016 which were not profitable but got the **highest IMDB-score**?

```python
# find movies with highest imdb_score
query2 = '''SELECT movie_title, imdb_score
            FROM movies
            WHERE \"Release Date\" >= '2016-01-01' AND \"Release Date\" < '2017-01-01' AND \"Worldwide Gross\" - \"Production Budget\" < 0
            ORDER BY imdb_score;
'''
pd.read_sql(query2, conn).head()
```

```python
# find movies with highest imdb_score (descending order)
query2 = '''SELECT movie_title, imdb_score
            FROM movies
            WHERE \"Release Date\" >= '2016-01-01' AND \"Release Date\" < '2017-01-01' AND \"Worldwide Gross\" - \"Production Budget\" < 0
            ORDER BY imdb_score DESC;
'''
pd.read_sql(query2, conn) 
```

### 3. What is the worldwide gross revenue per weekday for Action movies for which the number of Facebook likes is known (i.e. is different than None)?

```python
# STEP BY STEP SOLUTION
```

What is the **worldwide gross revenue per weekday** for Action movies for which the number of Facebook likes is known (i.e. is different than None)?

```python
# movie table
query3 = '''SELECT *
            FROM movies;
''' 
pd.read_sql(query3, conn).head()
```

```python
# daily table
query3 = '''SELECT *
            FROM daily;
''' 
pd.read_sql(query3, conn).tail()
```

```python
# join tables
query3 = '''SELECT *
            FROM movies m, daily d
            WHERE m.mid = d.movie
''' 
pd.read_sql(query3, conn).tail() # 166121 records! Why is that? Is that the same as in the Daily table? Why?
```

What is the **worldwide gross revenue per weekday** for Action movies for which the number of Facebook likes is known (i.e. is different than None)?

```python
# sum by weekday
query3 = '''SELECT SUM(\"Worldwide Gross\"), weekday
            FROM movies m, daily d
            WHERE m.mid = d.movie 
            GROUP BY weekday;
'''
pd.read_sql(query3, conn)
```

What is the worldwide gross revenue per weekday **for Action movies** for which the number of Facebook likes is known (i.e. is different than None)?

```python
query3 = '''SELECT SUM(\"Worldwide Gross\"), weekday
            FROM movies m, daily d
            WHERE m.mid = d.movie AND genre="Action" 
            GROUP BY weekday;
'''
pd.read_sql(query3, conn)
```

What is the worldwide gross revenue per weekday for Action movies for which the **number of Facebook likes is known (i.e. is different than None)**?

```python
query3 = '''SELECT movie_facebook_likes
            FROM movies;
'''
pd.read_sql(query3, conn)
```

```python
query3 = '''SELECT movie_facebook_likes
            FROM movies
            WHERE movie_facebook_likes != "None"
'''
pd.read_sql(query3, conn)
```

```python
# alternatively
query3 = '''SELECT movie_facebook_likes
            FROM movies
            WHERE movie_facebook_likes NOT NULL
'''
pd.read_sql(query3, conn)
```

```python
# order results by Worldwide Gross (without SUM) does NOT give the desired result:
query3 = '''SELECT SUM(\"Worldwide Gross\"), weekday
            FROM movies m, daily d
            WHERE m.mid = d.movie AND genre="Action" AND movie_facebook_likes NOT NULL
            GROUP BY weekday
            ORDER BY \"Worldwide Gross\";
'''
pd.read_sql(query3, conn)
```

```python
# order results by SUM(Worldwide Gross) does
query3 = '''SELECT SUM(\"Worldwide Gross\"), weekday
            FROM movies m, daily d
            WHERE m.mid = d.movie AND genre="Action" AND movie_facebook_likes NOT NULL
            GROUP BY weekday
            ORDER BY SUM(\"Worldwide Gross\");
'''
pd.read_sql(query3, conn)
```

```python
# change ordering from ascending to descending
query3 = '''SELECT SUM(\"Worldwide Gross\"), weekday
            FROM movies m, daily d
            WHERE m.mid = d.movie AND genre="Action" AND movie_facebook_likes NOT NULL
            GROUP BY weekday
            ORDER BY SUM(\"Worldwide Gross\") DESC;
'''
pd.read_sql(query3, conn)
```

```python
# change column name (alias)
query3 = '''SELECT SUM(\"Worldwide Gross\") AS 'Worldwide Gross Revenue', weekday
            FROM movies m, daily d
            WHERE m.mid = d.movie AND genre="Action" AND movie_facebook_likes NOT NULL
            GROUP BY weekday
            ORDER BY SUM(\"Worldwide Gross\") DESC;
'''
pd.read_sql(query3, conn)
```
