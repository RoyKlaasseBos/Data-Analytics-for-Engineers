## 2. Plotting a Pandas Series

## 3. Plotting a Pandas DataFrame

```python
# a short introduction to dictionaries
student_grades_list = [6.7, 8.2, 9.0]
student_grades_dict = {"Sem": 6.7, "Lucas": 8.2, "Tess": 9.0}

# why is it called a "dictionary"? 
# what is the advantage compared to lists?  
```

```python
# pass a dictionary a key to get its value
student_grades_dict["Sem"]
```

```python
# create a dataframe from two pandas series
student_grades_a1 = pd.Series([6.7, 8.2, 9.0, 5.0, 3.4, 4.5, 7.0, 3.4, 8.4, 5.7])
student_grades_a2 = pd.Series([7.5, 8.2, 9.8, 4.9, 5.3, 4.4, 8.6, 7.4, 5.4, 6.7])

# combine both dataframes
student_grades_a1_and_a2 = pd.DataFrame({"a1": student_grades_a1, "a2": student_grades_a2})

# what is the difference between a pandas DataFrame and a Series?
```

```python
student_grades_a1_and_a2.head()
```

```python
# visualize data points (basic plotting)
ax = student_grades_a1_and_a2.plot()
```

## 4. Adding elements to a plot

```python
# add title
ax = student_grades_a1_and_a2.plot()
ax.set_title("Grades assignment 1 and 2")
# why did we assign the title to "ax"?
```

How do we remove **`<matplotlib.text.Text at 0x11d9bd0f0>`** from the plot? 

```python
# add title (without matplotlib.text line)
ax = student_grades_a1_and_a2.plot()
ax.set_title("Grades assignment 1 and 2");
# what is the difference with the previous plot?
```

```python
# change size of title
ax = student_grades_a1_and_a2.plot()
ax.set_title("Grades assignment 1 and 2", size=14);
```

```python
# change weight of title
ax = student_grades_a1_and_a2.plot()
ax.set_title("Grades assignment 1 and 2", weight="bold");
# how could we change the weight AND the size?
```

```python
# there are many other properties (e.g. color)!
ax = student_grades_a1_and_a2.plot()
ax.set_title("Grades assignment 1 and 2", color="red"); # yellow, blue, etc.
# what if we want to use a different color? 
```

```python
# some other handy matplotlib chart elements
ax = student_grades_a1_and_a2.plot(color=["green", "red"], style="--", marker="o", figsize=(10,4))
ax.set_title("Grades assignment 1 and 2")
ax.set_ylabel("Grade") # label on y-axis (vertical)
ax.set_xticks(range(10)) # how many xticks? (for strings)
ax.set_xticklabels(["Sem", "Lucas", "Tess", "Adam", "James", "Max", "Fleur", "Zoë", "Luuk", "Noor"], size=9); # for string labels 
```

## 5. Different kinds of plots

```python
ax = student_grades_a1_and_a2.plot(kind='bar', color=["green", "red"])
```

```python
ax = student_grades_a1_and_a2.plot(kind='density', color=["green", "red"])
# what about the "11" and "12" grades? 
```

```python
ax = student_grades_a1_and_a2.plot(kind='box')
ax.set_title("Grade distribution assignment 1 and 2");
# what conclusion can we draw? 
```
